#Question : 01

Create a function that takes three parameters where:
    
 x is the start of the range (inclusive).

 y is the end of the range (inclusive).

 n is the divisor to be checked against.

Return an ordered list with numbers in the range that are divisible by the third parameter n.

Return an empty list if there are no numbers that are divisible by n.

Examples

list_operation(1, 10, 3) ➞ [3, 6, 9]

list_operation(7, 9, 2) ➞ [8]

list_operation(15, 20, 7) ➞ []

```python
#code
def value():
    try:
        num=int(input("Enter the number of test cases to be checked : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def values():
    result=[]
    try:
        x=int(input("Enter the start of the range : "))
        y=int(input("Enter the end of the range : "))
        n=int(input("Enter the divisor to be checked : "))
    except Exception as e:
        print(e)
        return values()+[]
    try:
        for i in range(x,y):
            if i%n==0:
                result.append(i)
            else:
                continue
    except Exception as e:
        print(e)
    return result
num=value()
for i in range(num):
    x=values()
    print(x)
```

#Question : 03

A group of friends have decided to start a secret society. The name will be the first letter of
each of their names, sorted in alphabetical order.

Create a function that takes in a list of names and returns the name of the secret society.

Examples

society_name(["Adam", "Sarah", "Malcolm"]) ➞ "AMS"

society_name(["Harry", "Newt", "Luna", "Cho"]) ➞ "CHLN"

society_name(["Phoebe", "Chandler", "Rachel", "Ross", "Monica", "Joey"])

```python
#code
def value():
    try :
        num=int(input("Enter the number of elements in the list : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def secret():
    num=value()
    N=[]
    final=""
    for i in range(num):
        x=input("Enter the name of the SS member : ")
        N.append(x)
    N.sort()
    for i in range(len(N)):
        p=N[i][0]
        final=final+p
    return final
num=int(input("Enter the number of test to be performed : "))
for i in range(num):
    m=secret()
    print(m)           
```

#Question : 04

An isogram is a word that has no duplicate letters. Create a function that takes a string and
returns either True or False depending on whether or not it&#39;s an "isogram".

Examples

is_isogram("Algorism") ➞ True

is_isogram("PasSword") ➞ False

*Not case sensitive.

is_isogram("Consecutive") ➞ False

Notes

 Ignore letter case (should not be case sensitive).

 All test cases contain valid one word strings.

```python
#code
def value():
    try:
        num=int(input("Enter the number of test cases : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def check():
    k=[]
    n=input("Enter the unidentified Isogram : ")
    N=n.lower()
    for i in N:
        if i not in k:
            k.append(i)
        else:
            return "False"
    else:
        return "True"
num=value()

for i in range(num):
    x=check()
    print(x)
```

#Question : 05

Create a function that takes a string and returns True or False, depending on whether the
characters are in order or not.

Examples

is_in_order("abc") ➞ True

is_in_order("edabit") ➞ False

is_in_order("123") ➞ True

is_in_order("xyzz") ➞ True

Notes

You don't have to handle empty strings.

```python
#code
def value():
    try:
        num=int(input("Enter the number of test cases : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def check():
    n=input("Enter the sequence to be checked : ")
    N=[]
    final=""
    for i in n:
        N.append(i)
    N.sort()
    for i in N:
        final=final+i
    if n==final:
        return True
    else:
        return False
num=value()
for i in range(num):
    x=check()
    print(x)
```
