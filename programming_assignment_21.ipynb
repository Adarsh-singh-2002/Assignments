#Question : 01

Write a function that takes a list and a number as arguments. Add the number to the end of
the list, then remove the first element of the list. The function should then return the updated
list.

Examples

next_in_line([5, 6, 7, 8, 9], 1) ➞ [6, 7, 8, 9, 1]

next_in_line([7, 6, 3, 23, 17], 10) ➞ [6, 3, 23, 17, 10]

next_in_line([1, 10, 20, 42 ], 6) ➞ [10, 20, 42, 6]

next_in_line([], 6) ➞ "No list has been selected"

```python
#code
def values():
    N=[]
    try:
        num=int(input("Enter the number of elements : "))
        if num==0:
            return " No list has been selected"
        for i in range(num):
            x=int(input())
            N.append(x)
        c=int(input("Enter the single argument : "))
        N.append(c)
        N.pop(0)
    except Exception as e:
        print(e)
        s=values()
        if isinstance(s,list):
            return s
        else:
            return []
    return N
num=int(input("Enter the number of test cases : "))
for i in range(num):
    N=values()
    print("Modified list is : ",N)

```

#Question : 03

Create a function that takes a string and returns a string with its letters in alphabetical order.

Examples

alphabet_soup("hello") ➞ "ehllo"

alphabet_soup("edabit") ➞ "abdeit"

alphabet_soup("hacker") ➞ "acehkr"

alphabet_soup("geek") ➞ "eegk"

alphabet_soup("javascript") ➞ "aacijprstv"

```python
#code
def value():
    try:
        s=input("Enter the string : ")
        N=[]
        final=""
        for i in s:
            N.append(i)
        N.sort()
        for i in N:
            final=final+i
    except Exception as e:
        print(e)
        return value()+""
    return final
num=int(input("Enter the number of tests : "))
for i in range(num):
    x=value()
    print(x)
```

#Question : 05

Write a function that takes a list of elements and returns only the integers.

Examples

return_only_integer([9, 2, "space", "car", "lion", 16]) ➞ [9, 2, 16]

return_only_integer(["hello", 81, "basketball", 123, "fox"]) ➞ [81, 123]

return_only_integer([10, "121", 56, 20, "car", 3, "lion"]) ➞ [10, 56, 20,
3]

return_only_integer(["String", True, 3.3, 1]) ➞ [1]

```python
#code
def value():
    try:
        num=int(input("Enter the number of tests to be done : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def values():
    try:
        N=[]
        n=int(input("Enter the number of values to be present in the list : "))
        for i in range(n):
            x=input(f"Enter the {i+1} element : ")
            N.append(x)
    except Exception as e:
        print(e)
        return values()+[]
    return N
num=value()
def integers(N):
    final=[]
    for i in N:
        if i.isdigit():
            final.append(int(i))
        else:
            continue
    return final
for i in range(num):
    x=values()
    final=integers(x)
    print(final)
    
```
