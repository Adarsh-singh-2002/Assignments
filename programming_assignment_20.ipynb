#Question : 01

Create a function that takes a list of strings and integers, and filters out the list so that it
returns a list of integers only.

Examples

filter_list([1, 2, 3, "a", "b", 4]) ➞ [1, 2, 3, 4]

filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) ➞ [0, 1729]

filter_list(["Nothing", "here"]) ➞ []

```python
#code
def value():
    try:
        num=int(input("Enter the number of test cases : "))
    except Exception as e:
        print(e)
        return value()+0
    return num
def values():
    try:
        N=int(input("Enter the number of values in this list : "))
        n=[]
        for i in range(N):
            x=input("Enter the value of element : "+str(i+1)+" ")
            n.append(x)
        return n
    except Exception as e:
        print(e)
        x=values()
        if isinstance(x,list):
            return x
        else:
            return []
num=value()
def filter_list_(num):
    for i in range(num):
        N=values()
        final=[]
        for i in N:
            if i.isdigit():
                final.append(int(i))
        print(final)
    return 
filter_list_(num)
```
